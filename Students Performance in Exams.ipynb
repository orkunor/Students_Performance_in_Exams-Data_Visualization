```python
# Let's create a new column in the form of overallpassstatus and examine whether it has passed or not.
# This column will look at whether a student has failed all of the courses.
df["Overall_PassStatus"] = df.apply(lambda x: "F" if x["Math_PassStatus"]== 'F' 
                                     or x["Reading_PassStatus"]== 'F' or x["Writing_PassStatus"]== 'F'
                                    else 'P' , axis =1)
df.Overall_PassStatus.value_counts() #Now we're looking at the number of them.
# if fully successful 812
```

```python
# Now let's find the average of 3 grades
df["Total Max"] = df["math score"]+df["reading score"] + df["writing score"]
```

```python
df["Percentage"] = df["Total Max"]/3
```

```python
df["Percentage"] = df["Total Max"]/3
```

```python
p=sns.countplot(x="Percentage",data=df,palette="muted") 
plt.xticks(rotation=90)
plt.show()
```

```python
def GetGrade(Percentage,Overall_PassStatus):
    if (Overall_PassStatus == 'F'):
        return 'F'
    if (Percentage >= 80):
        return 'A'
    if (Percentage >= 70):
        return 'B'
    if (Percentage >= 60):
        return 'c'
    if (Percentage >= 50):
        return 'D'
    if (Percentage >= 40):
        return 'E'
    else:
        return 'F'
df['Grade'] = df.apply(lambda x : GetGrade(x['Percentage'],x['Overall_PassStatus']),axis=1)

df.Grade.value_counts()
    
```

```python
# reach the numbers of boys and girls, let's visualize them too
df_gender=df["gender"]
female_count = df_gender.value_counts()["female"]
male_count = df_gender.value_counts()["male"]
```

```python
female_count , male_count
```

```python
female_ratio = female_count/(male_count+female_count) * 100
male_ratio = male_count/(female_count+male_count) * 100
```

```python
print("Number of Girls Ratio : ",int(female_ratio))
print("Number of men Ratio :",int(male_ratio))
```

```python
labels = ["Girl","Man"]
sizes=[female_ratio,male_ratio]
colors=["pink","lightblue"]
plt.pie(sizes,labels=labels,colors=colors,autopct="%1.1f%%",startangle=90)

plt.axis("equal")
plt.title("Girl-Male Ratios")
plt.show()
```

```python
# now let's just see the girls' success in math class
df_female=df[df["gender"]=="female"]
df_male=df[df["gender"]=="male"]
female_math_mean=df_female["math score"].mean()
male_math_mean=df_male["math score"].mean()

female_to_male_math_ratio = female_math_mean/male_math_mean

print("Ratio of girls to boys : ",female_to_male_math_ratio)
```

```python
# now let's show the averages on the graph 
# Let's create a bar chart
labels=["Girls","men"]
values=[female_math_mean,male_math_mean]
colors=["pink","lightblue"]
plt.bar(labels,values,color=colors) # here we create our graph  

plt.axhline(y=male_math_mean,color="gray",linestyle="--") # we create a horizontal line to see it exactly 
plt.text(-0,male_math_mean,"Average(Men)",ha="right")

plt.axhline(y=female_math_mean,color="gray",linestyle="--")
plt.text(1.1,female_math_mean,"Average(Girls)",ha="left")
plt.text(-0.8,female_math_mean*1.05,"Oran: {:.2f}".format(female_to_male_math_ratio),ha="center")

plt.xlabel("cinsiyet")
plt.ylabel("mat.notu")
plt.title("KÄ±zlarÄ±n Mat Dersindeki BaÅŸarÄ±sÄ±nÄ±n Erkeklere OranÄ±")
# As can be seen, the average for men is generally higher
```

```python
# Now let's calculate the ratio of the education level of the families to the mathematics or other subjects.
df.head()
```

```python
parental_edu_counts=df["parental level of education"].value_counts() # we got the number

# now let's convert the education levels into a list

parental_edu_levels = parental_edu_counts.index.tolist()

math_scores=[]

for level in parental_edu_levels:
    math_scores.append(df[df["parental level of education"]==level]["math score"].mean())
    
plt.bar(parental_edu_levels,math_scores)

plt.xlabel("family education level")
plt.ylabel("math average")
plt.title("family education math relationship")
plt.xticks(rotation=90)
plt.show()
```

```python
parental_edu_counts=df["parental level of education"].value_counts() 

parental_edu_levels = parental_edu_counts.index.tolist()

reading_scores=[]

for level in parental_edu_levels:
    reading_scores.append(df[df["parental level of education"]==level]["reading score"].mean())
    
plt.bar(parental_edu_levels,math_scores)

plt.xlabel("family education level")
plt.ylabel("reading skill")
plt.title("family education relationship")
plt.xticks(rotation=90)
plt.show()
```

```python
# The lunch column doesn't have much of an effect, but on the notes that will have an effect, as ethnicity has different variants.
# Let's examine the effect of ethnicity on mathematics 
ethnic=df.groupby("race/ethnicity")["math score"].mean().reset_index()
plt.bar(ethnic["race/ethnicity"],ethnic["math score"])
plt.title("Mathematics grade by ethnicity")
plt.xlabel("Ethnicity")
plt.ylabel("Maths")
plt.show()
```

```python
# Let's evaluate math grades by grouping them according to ethnicity and parental education levels.

ethnic=df.groupby(["race/ethnicity","parental level of education"])["math score"].mean().reset_index()
sns.catplot(x="parental level of education",y="math score",hue="race/ethnicity",kind="bar",data=ethnic,height=6,aspect=1.5)
# We set the aspect ratio of the aspect figure!!!
plt.title("Mathematics grade by ethnicity")
plt.xlabel("Ethnicity")
plt.ylabel("Math")
plt.show()
```
